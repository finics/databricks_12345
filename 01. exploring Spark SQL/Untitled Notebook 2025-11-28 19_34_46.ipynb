```python
df = spark.read.json(
    path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/dataset/product_Information_001.json",
    multiLine = True,
)
df.printSchema()
```

```python
from pyspark.sql.functions import col
df.select(col("name"),col("details.screen.size"))
```

```python
df_selected = df.select(
    col("product_id"),
    col("name"),
    col("details.screen.size").alias("screensize"),
    col("details.storage.capacity").alias("storage_capacity")
)
df_selected.display()
```
