```python
df = spark.read.csv(path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/dataset/employee.csv",
                    sep="|",
                    inferSchema=True,
                    header=True,
                    quote="'")
display(df)  
```

##Select Op

```python
df.select("id","name").display()
```

```python
from pyspark.sql.functions import col
df.select(col("id").alias("user_id"),col("name")).display()
```

```python
df.filter(col("desig") == "Developer").display()
### df.filter(col("desig") == "Manager").select(col("id").alias("user_id"))
```

```python
df.filter(col("desig") == "Developer" & (col("gen") == "M")).select(col("id").alias("user_id")).display()
```

```python
df.filter(col("exp").between(1,6) & (col("gen") == "M")).select(col("name")).display()
```

```python
grouped_df = df.groupBy("exp","gen").agg({"exp":"avg","id":"count"})
display(grouped_df)
df.groupBy("desig").count
```

```python
#df.groupBy("gen").count()
df.groupBy("Desig").count().orderBy(col("count")).display()
```

```python

df.groupBy("Desig").agg({"exp":"min"}).display()
```

```python
from pyspark.sql.functions import lit
#df.withColumn("salary",col("col_current_expected_salary").cast("int")).display()
df.withColumn("is_emplyed", lit("Yes")).display()
```

```python
from pyspark.sql.functions import when

df.withColumn(
    "catagory", 
    when(col("exp") < 3, "Junior").
    when(col("exp") >= 3, "Lead").
    when(col("exp") >= 6, "Senior").
    otherwise("Invalid Experience")

).display()
```

```python
from pyspark.sql.functions import col
from pyspark.sql.functions import when
df = df.withColumn("id", when(col("id").isNull(), -1).otherwise(col("id"))) \
       .withColumn("exp", when(col("exp").isNull(), 0).otherwise(col("exp")))
df.display()
#df.filter(col("exp").isNotNULL() & (col("gen") == "M")).select(col("name")).display()
```
