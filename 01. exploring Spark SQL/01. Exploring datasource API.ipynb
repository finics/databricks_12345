###Custom Schema

```python
from pyspark.sql.types import *

USER_SCHEMA = StructType(
    [
        StructField("id", IntegerType(), True),
        StructField("age", IntegerType(), True),
        StructField("gen", StringType(), True),
        StructField("designation", StringType(), True),
        StructField("salary", IntegerType(), True),
    ]
)
df = spark.read.csv(
    path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/dataset/sample_dataset/users_001.csv",
    header=True,
    schema = USER_SCHEMA
)
df.display()
```

#Convert CSV to Json
Overwrite, Append, Ignore mode while saving

```python
df = spark.read.csv(
    path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/dataset/sample_dataset/users_001.csv",
    header=True,
    inferSchema=True
)
df.write.save(format="json", path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/dataset/output_json",mode ="overwrite")
```
