## Create Delta table

```python
%sql 

CREATE TABLE quickstart_catalog.quickstart_schema.users_int(
  id INT,
  name STRING,
  dob DATE,
  email STRING,
  gender STRING,
  country STRING,
  region STRING,
  city STRING,
  asset INT,
  marital_status STRING
) USING DELTA;
DESCRIBE EXTENDED quickstart_catalog.quickstart_schema.users_int;
```

```python
%sql
COPY INTO quickstart_catalog.quickstart_schema.users_int FROM (
  SELECT
    id::INT,
    name::STRING,
    to_date(dob, 'yyyy-MM-dd') as dob,
    email::STRING,
    gender::STRING,
    country::STRING,
    region::STRING,
    city::STRING,
    asset::INT,
    marital_status::STRING,
    education::STRING FROM
  '/Volumes/quickstart_catalog/quickstart_schema/sandbox/dataset/users_staging/'
) FILEFORMAT = CSV FORMAT_OPTIONS ('header' = 'true')
```

```python
spark.sql("DROP TABLE IF EXISTS quickstart_catalog.quickstart_schema.users_int")
```
